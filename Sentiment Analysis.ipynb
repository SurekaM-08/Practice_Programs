```python
positive_words = ["good", "great", "happy", "love", "excellent", "nice", "amazing", "awesome"]
negative_words = ["bad", "sad", "hate", "poor", "terrible", "worst", "awful", "ugly"]

sentence = input("Enter a sentence: ").lower()

positive_count = 0
negative_count = 0

words = sentence.split()

for word in words:
    if word in positive_words:
        positive_count += 1
    elif word in negative_words:
        negative_count += 1

print(f"\nPositive words: {positive_count}")
print(f"Negative words: {negative_count}")

if positive_count > negative_count:
    print("Sentiment: Positive 😊")
elif negative_count > positive_count:
    print("Sentiment: Negative 😞")
else:
    print("Sentiment: Neutral 😐")
```
