```python
def chatbot():
    print("=" * 50)
    print("🤖 Welcome! I'm your AI Assistant")
    print("Type 'bye' to exit")
    print("=" * 50)
    
    # Responses grouped by topic
    responses = {
        "greeting": ["Hello! How can I help you today?", "Hi there! What's up?"],
        "how_are_you": ["I'm doing great, thanks for asking! 😊", "All good! Ready to chat!"],
        "name": ["I'm your AI Chatbot! Nice to meet you!"],
        "creator": ["I was created with Python - pretty cool, right?"],
        "python": ["Python is awesome! It's a language that's easy to read and very powerful."],
        "ml": ["ML is how computers learn from data without being explicitly programmed. Cool stuff!"],
        "confused": ["Hmm, I didn't quite catch that. Can you rephrase?", "Sorry! I'm still learning. Try something else?"]
    }
    
    keywords = {
        "greeting": ["hello", "hi", "hey", "hiya"],
        "how_are_you": ["how are you", "how are ya"],
        "name": ["what is your name", "who are you"],
        "creator": ["who created you", "where do you come from"],
        "python": ["what is python", "tell me about python"],
        "ml": ["machine learning", "what is ml"],
    }
    
    while True:
        user = input("\n📝 You: ").lower().strip()
        
        if user == "bye":
            print("\n✨ Bot: Thanks for chatting! See you later! 👋")
            break
        
        # Check which category the input matches
        found = False
        for category, words in keywords.items():
            if any(word in user for word in words):
                import random
                print(f"🤖 Bot:{random.choice(responses[category])}")
                found = True
                break
        
        if not found:
            import random
            print(f"🤖 Bot: {random.choice(responses['confused'])}")
chatbot()
```
