```python
# 1. Check whether a number is Prime or Not
n = int(input("Enter a number: "))
is_prime = True
if n < 2:
    is_prime = False
for i in range(2, n):
    if n % i == 0:
        is_prime = False
if is_prime:
    print(n, "is a Prime number")
else:
    print(n, "is Not a Prime number")
```

```python
# 2. Find the Factorial of a number
n = int(input("Enter a number: "))
fact = 1
for i in range(1, n+1):
    fact = fact * i
print("Factorial of", n, "=", fact)
```

```python
# 3. Generate Fibonacci Series up to N terms
n = int(input("Enter number of terms: "))
a = 0
b = 1
print("Fibonacci Series:")
for i in range(n):
    print(a, end=" ")
    c = a + b
    a = b
    b = c
 
```

```python
# 4. Check whether a number is a Palindrome
n = int(input("Enter a number: "))
original = n
reverse = 0
while n > 0:
    digit = n % 10
    reverse = reverse * 10 + digit
    n = n // 10
if original == reverse:
    print(original, "is a Palindrome")
else:
    print(original, "is Not a Palindrome")
```

```python
# 5. Find the Sum of Digits of a given number
n = int(input("Enter a number: "))
total = 0
while n > 0:
    digit = n % 10
    total = total + digit
    n = n // 10
print("Sum of digits =", total)
```

```python
# 6. Reverse a number without using string functions
n = int(input("Enter a number: "))
reverse = 0
while n > 0:
    digit = n % 10
    reverse = reverse * 10 + digit
    n = n // 10
print("Reversed number =", reverse)
```

```python
# 7. Check whether a year is a Leap Year
year = int(input("Enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, "is a Leap Year")
else:
    print(year, "is Not a Leap Year")
```

```python
# 8. Find the Largest of Three Numbers
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))
if a >= b and a >= c:
    print("Largest =", a)
elif b >= a and b >= c:
    print("Largest =", b)
else:
    print("Largest =", c)
```

```python
# 9. Swap two numbers without using a third variable
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("Before swap: a =", a, ", b =", b)
a = a + b
b = a - b
a = a - b
print("After swap:  a =", a, ", b =", b)
```

```python
# 10. Count the number of digits in a given number
n = int(input("Enter a number: "))
count = 0
temp = n
while temp > 0:
    count = count + 1
    temp = temp // 10
print("Number of digits in", n, "=", count)
```

```python
# 11. Reverse a String without using slicing
s = input("Enter a string: ")
result = ""
for char in s:
    result = char + result
print("Reversed string =", result)
```

```python
# 12. Check whether a String is a Palindrome
s = input("Enter a string: ")
reverse = ""
for char in s:
    reverse = char + reverse
if s == reverse:
    print(s, "is a Palindrome")
else:
    print(s, "is Not a Palindrome")
```

```python
# 13. Count the number of Vowels and Consonants in a String
s = input("Enter a string: ")
vowels = 0
consonants = 0
for char in s.lower():
    if char.isalpha():
        if char in "aeiou":
            vowels = vowels + 1
        else:
            consonants = consonants + 1
print("Vowels =", vowels)
print("Consonants =", consonants)
 
```

```python
# 14. Find the Frequency of each Character in a String
s = input("Enter a string: ")
freq = {}
for char in s:
    if char in freq:
        freq[char] = freq[char] + 1
    else:
        freq[char] = 1
for char in freq:
    print(char, "->", freq[char])
 
```

```python
# 15. Check whether two Strings are Anagrams
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")
if sorted(s1.lower()) == sorted(s2.lower()):
    print(s1, "and", s2, "are Anagrams")
else:
    print(s1, "and", s2, "are Not Anagrams")
```

```python
# 16. Find the First Non-Repeated Character in a String
s = input("Enter a string: ")
freq = {}
for char in s:
    if char in freq:
        freq[char] = freq[char] + 1
    else:
        freq[char] = 1
for char in s:
    if freq[char] == 1:
        print("First non-repeated character =", char)
        break
```

```python
# 17. Remove all Spaces and Find Duplicate Characters in a String
s = input("Enter a string: ")
 
# Remove spaces
result = ""
for char in s:
    if char != " ":
        result = result + char
print("After removing spaces =", result)
 
# Find duplicates
freq = {}
for char in s:
    if char in freq:
        freq[char] = freq[char] + 1
    else:
        freq[char] = 1
print("Duplicate characters:")
for char in freq:
    if freq[char] > 1:
        print(char)
```
